```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([1, 2, 3, 4, 5, 7, 8,9]).reshape(-1, 1)
y = np.array([3, 3.5, 4, 4.2, 6, 6.5, 7, 7.2])
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(
    x,y,test_size=0.25,random_state=42
)
```

```python
from sklearn.tree import DecisionTreeRegressor
dt_reg = DecisionTreeRegressor(random_state=42)
dt_reg.fit(x_train, y_train)
```

```python
X_grid=np.linspace(1,8,100).reshape(-1,1)
y_pred=dt_reg.predict(X_grid)
plt.scatter(x,y,color='blue')
plt.plot(X_grid,y_pred,color='red')
plt.xlabel("Decision Tree Regression")
plt.ylabel("Years of Experience")
plt.title("Salary (LPA)")
plt.show()
```

```python
dt_deep=DecisionTreeRegressor(max_depth=None)
dt_deep.fit(x_train,y_train)
y_pred_deep=dt_deep.predict(X_grid)
plt.scatter(x,y)
plt.plot(X_grid,y_pred_deep,color='green')
plt.title("Overfitting Tree")
plt.show()
```

```python
dt_controlled=DecisionTreeRegressor(max_depth=None)
dt_controlled.fit(x_train,y_train)
y_pred_conntrolled=dt_controlled.predict(X_grid)
plt.scatter(x,y)
plt.plot(X_grid,y_pred_conntrolled,color='orange')
plt.title("Controlled Tree(max_depth=3)")
plt.show()
```

```python
from sklearn.metrics import mean_squared_error,r2_score
y_test_pred_deep=dt_deep.predict(x_test)
y_test_pred_controlled=dt_controlled.predict(x_test)
print("Deep Tree Metrics")
print("MSE:",mean_squared_error(y_test,y_test_pred_deep))
print("R2:",r2_score(y_test,y_test_pred_deep))
print("\nControlled Tree Metrics")
print("MSE:",mean_squared_error(y_test,y_test_pred_controlled))
print("R2:",r2_score(y_test,y_test_pred_controlled))
```

```python
np.random.seed(42)
X=np.arange(1, 21).reshape(-1,1)
y=np.array([
    10,12,15,18,20,25,30,28,35,40,
    45,42,50,55,60,58,65,70,72,75
])
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(
    X,y,test_size=0.25,random_state=42
)
```

```python
from sklearn.tree import DecisionTreeRegressor
dt_reg = DecisionTreeRegressor(random_state=42)
dt_reg.fit(x_train, y_train)
```

```python
from sklearn.ensemble import RandomForestRegressor
rf=RandomForestRegressor(
    n_estimators=100,
    random_state=42
)
rf.fit(x_train,y_train)
```

```python
X_grid=np.linspace(1,20,200).reshape(-1,1)
y_pred=dt_reg.predict(X_grid)
plt.scatter(X,y,color='black')
plt.plot(X_grid,dt_reg.predict(X_grid),color='red',label='Decision Tree')
plt.plot(X_grid,rf.predict(X_grid),color='green',label='Random Forest')
plt.xlabel("Decision Tree Regression")
plt.ylabel("Years of Experience")
plt.title("Salary (LPA)")
plt.show()
```

```python
from sklearn.metrics import mean_squared_error,r2_score
rf_pred=rf.predict(x_test)
dt_pred=dt_reg.predict(x_test)
print("Deep Tree Metrics")
print("MSE:",mean_squared_error(y_test,dt_pred))
print("R2:",r2_score(y_test,dt_pred))
print("\Random Forest Regression")
print("MSE:",mean_squared_error(y_test,rf_pred))
print("R2:",r2_score(y_test,rf_pred))
```

```python
import pandas as pd

df = pd.read_csv(r"C:\Users\gurra\DecisionMaking\house_data.csv")
df.head()
```

```python
Q1 = df["price"].quantile(0.25)
Q3 = df["price"].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

df_clean = df[
    (df["price"] >= lower_bound) &
    (df["price"] <= upper_bound)
]

print("Original shape:", df.shape)
print("After outlier removal:", df_clean.shape)
```

```python
X = df_clean.drop(columns=["price"])
y = df_clean["price"]

X = pd.get_dummies(X, drop_first=True)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42
)
```

```python
from sklearn.tree import DecisionTreeRegressor

dt_reg = DecisionTreeRegressor(
    max_depth=6,
    random_state=42
)

dt_reg.fit(X_train, y_train)
```

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(
    n_estimators=200,
    max_depth=10,
    random_state=42
)

rf.fit(X_train, y_train)
```

```python
rf_pred = rf.predict(X_test)
dt_pred = dt_reg.predict(X_test)
mse = mean_squared_error(y_test, rf_pred)
print("Random Forest:")
print("Mean Squared Error:", mse)
r2 = r2_score(y_test, rf_pred)
print("R² Score:", r2)
print("Decision Tree:")
mse = mean_squared_error(y_test, dt_pred)
print("Mean Squared Error:", mse)
r2 = r2_score(y_test, dt_pred)
print("R² Score:", r2)
```

```python

feature_name = "sqft_living"

# Scatter plot of actual prices
plt.figure(figsize=(8,5))
plt.scatter(df_clean[feature_name], y, color='black', label='Actual Prices')

# Create grid (1D)
X_grid = np.linspace(
    df_clean[feature_name].min(),
    df_clean[feature_name].max(),
    200
)

# Create DataFrame for predictions (all other features fixed at mean)
X_grid_df = pd.DataFrame(
    np.tile(X.mean().values, (200, 1)),
    columns=X.columns
)

# Replace feature with grid values
X_grid_df[feature_name] = X_grid

# Predictions
dt_pred = dt_reg.predict(X_grid_df)
rf_pred = rf.predict(X_grid_df)

# Plot predictions
plt.plot(X_grid, dt_pred, color='red', label='Decision Tree')
plt.plot(X_grid, rf_pred, color='green', label='Random Forest')

plt.xlabel("Living Area (sqft)")
plt.ylabel("House Price")
plt.title("Decision Tree vs Random Forest – House Price Prediction")
plt.legend()
plt.show()
```

```python
X=np.array([1,2,3,4,5,6]).reshape(-1,1)
y=np.array([35,40,50,60,65,70])
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
lin_reg = LinearRegression()
lin_reg.fit(X_train, y_train)
y_train_pred=lin_reg.predict(X_train)
y_test_pred=lin_reg.predict(X_test)
```

```python
print("GOOD FIT (Linear Regression)")
print("Train MSE:",mean_squared_error(y_train,y_train_pred))
print("Train R2:",r2_score(y_train,y_train_pred))
```

```python
plt.scatter(X,y)
plt.plot(X,lin_reg.predict(X),color='green')
plt.title("Good Fit: Marks vs Study Hours")
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.show()
```

```python
poly = PolynomialFeatures(degree=5)

X_train_poly = poly.fit_transform(X_train)
X_test_poly = poly.transform(X_test)

overfit_model=LinearRegression()
overfit_model.fit(X_train_poly,y_train)
y_train_pred_poly=overfit_model.predict(X_train_poly)
y_test_pred_poly=overfit_model.predict(X_test_poly)
```

```python
print("\nOVERFITTING MODEL (Polynomial Degree=5)")
print("Train MSE:",mean_squared_error(y_train,y_train_pred_poly))
print("Train R2:",r2_score(y_train,y_train_pred_poly))
```

```python
X_grid = np.linspace(1, 6, 100).reshape(-1, 1)
X_grid_poly = poly.transform(X_grid)

plt.scatter(X, y)
plt.plot(X_grid, overfit_model.predict(X_grid_poly), color='red')
plt.title("Overfitting: Memorizing Study Hours")
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.show()
```

```python
poly = PolynomialFeatures(degree=5)
X_poly = poly.fit_transform(X)
```

```python
from sklearn.linear_model import Ridge
ridge=Ridge(alpha=1)
ridge.fit(X_poly,y)
y_ridge_pred=ridge.predict(X_poly)
```

```python
from sklearn.linear_model import Lasso
lasso=Lasso(alpha=0.1)
lasso.fit(X_poly,y)
y_lasso_pred=lasso.predict(X_poly)
```
